```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, sosfreqz

# Paramètres du filtre passe-bande
lowcut = 1500.0  # Fréquence basse de coupure
highcut = 8000.0  # Fréquence haute de coupure
samplerate = 44100  # Taux d'échantillonnage
order = 6  # Ordre du filtre

# Création du filtre passe-bande
sos = butter(order, [lowcut], analog=False, btype='lowpass', fs=samplerate, output='sos')

# Calculer la réponse en fréquence du filtre
w, h = sosfreqz(sos, worN=2000, fs=samplerate)

# Tracer la réponse en fréquence
plt.figure(figsize=(8, 6))
plt.plot(w, 20 * np.log10(np.maximum(abs(h), 1e-5)), label="Filtre Passe-Bande (1500Hz - 8000Hz)")
plt.title('Réponse en fréquence du filtre passe-bande')
plt.xlabel('Fréquence [Hz]')
plt.ylabel('Amplitude [dB]')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```
